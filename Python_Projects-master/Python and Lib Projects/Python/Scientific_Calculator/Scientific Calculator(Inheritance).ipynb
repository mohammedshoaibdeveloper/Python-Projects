# Inheritance Example
  Scientific Calculator  drived by simple calculator

```python
import math
class Calculator():
    def __init__(self,a,b):
        self.num1=a
        self.num2=b
        
    def Add(self):
        print('It is Scientific calculator')
        print("===========================")
            
        print("Addition of Two numbers")
        print("=======================")
        add=self.num1+self.num2
        print(f"Addition of two number is:{add}\n")
        
    def Sub(self):
        print("Subtraction of Two numbers")
        print("========================\n")
        sub=self.num1 - self.num2
        print(f"Subtraction of Two number is: {sub}\n")
        
        
    def Multiply(self):
        print("Multiplication of Two numbers")
        print("===========================\n")
        Mul=self.num1 * self.num2
        print(f"Multiplication of Two number is: {Mul}\n")
        
        
    def Div(self):
        print("Division of Two of Two numbers")
        print("============================\n")
        Division=self.num1 / self.num2
        print(f"Divison of Two number is: {Division}\n")
        
    def Avg(self):
        print("Average of Two numbers")
        print("=======================")
        avg=self.num1+self.num2/2
        print(f"Addition of Two number is: {avg}\n")
        
#child Class....

class Scientific_calculator(Calculator):
   
    def __init__(self,a,b):
        super().__init__(a,b)
          
    def square(self,sq):
        print("Square of Numbers")
        print("=================")
        self.sq=sq
        square=self.sq**2
        print(f"square Root of the Following no is: {square}\n")

        
    def Square_Root(self,sq_r):
        print("Square Root of Numbers")
        print("======================")
        self.sq_root=sq_r
        square_Root=math.sqrt(self.sq_root)
        print(f"square Root of the Following no is: {square_Root}\n")

        
    def Trignometric_Funtions(self):
        print("Trignometric Functions")
        print("======================")
        
        print("Which Function would you like to performed!")
        trignometric_functions=['sin','cos','tan']
        for i in trignometric_functions:
            print("-",i)
            
    def Cos(self,cos):
        self.cos_an=cos
        f1=math.cos(self.cos_an)
        print(f"Cos {self.cos_an} is {f1}\n")
        
    def Sin(self,sin):
        self.sin_an=sin
        f2=math.sin(self.sin_an)
        print(f"Sin {self.sin_an} is {f2}\n")

    def tan(self,tan):
        self.tan_an=tan
        f3=math.tan(self.tan_an)
        print(f" tan {self.tan_an} is {f3}\n")
```

```python
a=int(input("Enter 1st Number : "))
b=int(input("Enter 2nd Number : "))
c=Scientific_calculator(a,b)

c.Add()
c.Sub()
c.Multiply()
c.Div()
c.Avg()

s=int(input("Enter a number for square : "))
c.square(s)

sq=int(input("Entr a number for square root : "))
c.Square_Root(sq)

c.Trignometric_Funtions()

cos=int(input("Enter a Angle for cos number into radian"))
c.Cos(cos)

tan=int(input("Enter a Angle for tan number into radian"))
c.tan(tan)

sin=int(input("Enter a Angle for sin number into radian"))
c.Sin(sin)
```
