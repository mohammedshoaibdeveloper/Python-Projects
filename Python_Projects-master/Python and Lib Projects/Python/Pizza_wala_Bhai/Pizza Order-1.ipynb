```python
Available_Size=['Medium',"Large","Small","Extra Large"]
Available_Flavours = ["Fjita", "Tikka", "Mexican",'vegitable']
Available_Toppings=['Olives', 'Chesse','Hurbs']
print("Hello..! I'm from PIzza Corneer")
print("How can i Help You ?")
customers=input("Have you Ordered First Time?").lower()
if customers=="yes":
    print("==============")
    print("==============")
    print("Avalible Menu")
    print("==============")
    print("==============")
    
    print("\n")
    print("Avalible Size")
    print("==============\n")
    for i in Available_Size:
        print("-",i)
        
        
    print("\n")
    print("Avalible Flavors")
    print("================\n")
    for i in Available_Flavours:
        print("-",i)
    print("\n")
    
    print("Avalible TOppings")
    print("=================")
    for i in Available_Toppings:
        print("-",i)

pizza_Order=True
while pizza_Order:
    
    c_Available_Size=input("Enter your Size of Pizza: ")
    if c_Available_Size in Available_Size:
        print("Thanks You !")
        pizza_Order=False
       
    elif c_Available_Size not in Available_Size:
            c_Available_Size=input("Not Avaliable ! Enter your Size  again of Pizza that Given on Above in the menu: ")
            if c_Available_Size in Available_Size:
                    print("Thanks You !")
                    pizza_Order=False
    #flavor
    
    c_Available_flavor=input("Enter your Flavor of Pizza: ")
    if c_Available_flavor in Available_Flavours:
        print("Thanks You !")
        pizza_Order=False
       
    elif c_Available_flavor not in Available_Flavours:
            c_Available_flavor=input("Not Avaliable ! Enter your Flavor  again of Pizza that Given on Above in the menu: ")
            if c_Available_flavor in Available_Flavours:
                    print("Thanks You")
                    pizza_Order=False
                    
    #topping
    c_Available_topping=input("Enter your Topping of Pizza: ")
    if c_Available_topping in Available_Toppings:
        print("Thanks You, Your Order has been done! !")
        print(f"- Your pizza size is: {c_Available_Size}\n-Your Flavor is {c_Available_flavor}\n-Your Topping is {c_Available_topping}")
        
        pizza_Order=False
       
    elif c_Available_topping not in Available_Toppings:
            c_Available_topping=input("Not Avaliable ! Enter your Topping  again of Pizza that Given on Above in the menu: ")
            if c_Available_topping in Available_Toppings:
                    print("Thanks You !")
                    pizza_Order=False
   

          
```
